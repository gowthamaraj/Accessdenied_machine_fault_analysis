```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

min= cv2.imread('capture.jpg')
#min = cv2.medianBlur(min,5)
min = cv2.cvtColor(min,cv2.COLOR_BGR2RGB)


tre1 = cv2.imread('contour.jpg')
tre1 = cv2.cvtColor(tre1,cv2.COLOR_BGR2RGB)
tre1 = cv2.medianBlur(tre1,7)
    
my_method= eval('cv2.TM_CCORR_NORMED')
res1 = cv2.matchTemplate(tre1,min,my_method)

#plt.imshow(res1)


retval, threshold = cv2.threshold(res1, 222, 255, cv2.THRESH_TRUNC)
plt.imshow(threshold,cmap='gray')
```
